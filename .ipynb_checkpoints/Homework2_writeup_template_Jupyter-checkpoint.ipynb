### Part b

The error in the numbers occurs because of how computers represent numbers. They using the floating point system, meaning that you take some binary number (which is the sum of various 2^n terms, n being negative integers, e.g. 0.5 + 0.25 + 0.0625.....) and then multiply by 2^exponent. This can get many numbers. It can get every integer up to a certain point and it can get anything of a power of 2. However, it cannot perfectly represent anything that isn't a sum of powers of 2, such as 0.1. One tenth can be represented as .0625 + .03125 + ... relatively accurately, but it's just off by the slightest bit. So when you add this number many times to itself, the small amount of error becomes larger and larger. When you do it 100_000 times, you get an error of x1, and when you do it 100_000_000 times, you get an error of x2. The error of x2 is larger because you're literally taking a small amount of error and doing it more times. However, the error in x2 is bigger by a scale of almost 10^6 compared to x1, while they only do the operation more times by a scale of 10^3. This problem shows another issue fith the floating point system, which is that the computer struggles to represent very large numbers accurately. Additionally, x3 and x4 can be seen to have absolutely no error because they are the sums of exact powers of 2: .25 and .5. This means the floating point system has very little trouble representing the numbers and adding them together. Of course the .1 in x1 and x2 lead to a non-zero error.

### Part c

As I said in part b, computers use the floating point system to store numbers. This means that it represents numbers as the sum of negative powers of 2 (e.g. 0.5 + 0.25 + 0.125 + 0.0625 ...) times some power of 2. This makes representing the sum of many .25s or .5s very easy for the computer. This is why x3 and x4 have no error. The computer has an easy time storing the exact value of x3 and x4; no error necessary.

## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1*np.pi, np.pi, 100)

ycos = np.zeros(100)
y1 = np.zeros(100)
y3 = np.zeros(100)
y14 = np.zeros(100)

for i in range(100):
    ycos[i] = np.cos(x[i])
    
    for k in range(1 + 1):
        y1[i] = y1[i] + np.power(-1, k) * np.power(x[i], 2*k) / np.math.factorial(2*k)
    
    for k in range(3 + 1):
        y3[i] = y3[i] + np.power(-1, k) * np.power(x[i], 2*k) / np.math.factorial(2*k)
        
    for k in range(14 + 1):
        y14[i] = y14[i] + np.power(-1, k) * np.power(x[i], 2*k) / np.math.factorial(2*k)


plt.plot(x,ycos,color='black', label="cos(x)", linewidth=2, linestyle='-')
plt.plot(x,y1,color='blue', label="n = 1", linewidth=2, linestyle='--')
plt.plot(x,y3,color='red', label="n = 3", linewidth=2, linestyle='-.')
plt.plot(x,y14,color='magenta', label="n = 14", linewidth=2, linestyle=':')
plt.xlabel('x-values', fontdict={"fontsize": 15})
plt.ylabel('cos(x) approximations', fontdict={"fontsize": 15})
plt.title("cos(x) and its Taylor approximations.", fontdict={"fontsize": 15})

plt.legend(loc="upper left")
plt.show()
```
